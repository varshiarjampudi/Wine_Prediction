```python
def log_trasform(col):
    return np.log1p(col[0])

wine_df["residual sugar"] = wine_df[["residual sugar"]].apply(log_trasform , axis=1)
wine_df["chlorides"] = wine_df[['chlorides']].apply(log_trasform , axis=1)
wine_df["total sulfur dioxide"] = wine_df[['total sulfur dioxide']].apply(log_trasform , axis=1)
wine_df["free sulfur dioxide"] = wine_df[['free sulfur dioxide']].apply(log_trasform , axis=1)
wine_df["density"] = wine_df[['density']].apply(log_trasform , axis=1)
```

```python
x = wine_df.drop('quality' , axis =1)
y = wine_df['quality']
```

```python
y.value_counts()
```

```python
#data balaancig using SMOTE
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors = 4)

x,y = oversample.fit_resample(x.fillna(0) , y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(
        x, y, test_size=0.2, random_state=42
    )
    model.fit(x_train, y_train)   #  MUST be y_train
    return model.score(x_test, y_test) * 100
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
accuracy_lr = classify(model, x, y)

print("LogisticRegression Accuracy {:.2f}%".format(accuracy_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc_model = DecisionTreeClassifier(random_state=42)
dtc_accuracy = classify(dtc_model, x, y)

print("Decision Tree Accuracy {:.2f}%".format(dtc_accuracy))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_accuracy = classify(rf_model, x, y)

print("Random Forest Accuracy {:.2f}%".format(rf_accuracy))
```

```python
#ploting the accuracies diffrent models
models = ['LogisticRegression' , 'DecisionTreeClassifier' , 'RandomForestClassifier']
accuracy = [accuracy_lr , dtc_accuracy, rf_accuracy]
plt.figure(figsize=(10,6))
sns.barplot(x=models,y=accuracy , palette='pastel')
plt.show()
```

```python
#preddict the quality of wine
#inserting new data

model = RandomForestClassifier()
model.fit(x,y)

sample_data = (7.4 , 0.7 , 0 , 1.9 ,0.076 , 11 , 34 ,0.9978 ,3.15 ,0.56 ,9.4 )
sample_data =np.array(sample_data).reshape(1,-1)
predection = model.predict(sample_data)
print("predection wine quality :" ,predection[0])
```
